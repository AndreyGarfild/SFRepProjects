Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

Как вы понимаете, нам необходимо выделить каждый параметр в отдельный столбец.

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

*Совет: обратите внимание на структуру текста в столбце, в части на то, как разделены параметры пола, возраста и даты рождения между собой - символом ' , '.

```python
# Функция получения пола
def get_gender(raw_str):
    str_splitted = raw_str.split(' , ')[0][0]
    return str_splitted

# Функция получения возраста
def get_age(raw_str):
    str_splitted = raw_str.split(' , ')[1].split(' ')[1]
    return str_splitted

# Заменяем старый признак новыми категориями
main_df["Пол"] = main_df["Пол, возраст"].apply(get_gender)
main_df["Возраст"] = main_df["Пол, возраст"].apply(get_age).astype(int)

# Проверка данных
print(f"Уникальные пол:{main_df['Пол'].unique()}")
print(f"Уникальные возраст: {main_df['Возраст'].unique()}\n")

# Вывод статистики

man_percentage = round(main_df[main_df["Пол"]=='М'].shape[0] / main_df.shape[0] * 100, 2)
woman_percentage = round(main_df[main_df["Пол"]=='Ж'].shape[0] / main_df.shape[0] * 100, 2)

print(f"Процент мужчин: {man_percentage}")
print(f"Процент женщин: {woman_percentage}\n")

print(f"Средний возраст соискателей: {round(main_df['Возраст'].mean(), 2)}")
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def check_and_convert(lst):
    try:
        if len(lst) > 4 and len(lst)<2: raise ValueError(f"Список содержит не правильное кол-во элементов: {len(lst)}")
        if len(lst) == 4: lst = lst[:3] 
        converted_list = []
        for item in lst:
       
            # Преобразование элемента в int
            int_item = int(item)
            
            # Проверка, что элемент больше нуля
            if int_item <= 0:
                raise ValueError(f"Элемент {item} не больше нуля.")
            
            converted_list.append(int_item)

    except ValueError as e:
        # Обработка ошибки, если преобразование не удалось или элемент не больше нуля
        print(f"Список не прошел валидацию или преобразование: {lst}\n Элемент:{item} \n Ошибка:{e}")
        

    return converted_list

def count_month_by_list(list):
    total_month = 0
    if list[1].startswith('л') or list[1].startswith('г'): list[1]=12
    elif list[1].startswith('м'): list[1]=1
    list = check_and_convert(list)

    total_month = list[0]*list[1]

    if len(list)==3: 
            total_month += list[2]   
        

    return total_month

def raw_stage_handle(raw_str):
    str_splitted = str(raw_str).split(' ')
    if  (str_splitted[0] == 'nan') or (str_splitted[0] == 'Не'): return np.NaN      # Проеряем значение на "пустоту" и заменяем на np.NaN
    
    stage_stripped = str_splitted[2:6]

    if not stage_stripped[-1].startswith('мес'): 
        stage_stripped = stage_stripped[:2]

    return count_month_by_list(stage_stripped)

main_df["Опыт работы"] = main_df["Опыт работы"].apply(raw_stage_handle)

# Блок вывода статистик
print(f"Медианный опыт работы в месяцах: {main_df['Опыт работы'].median()}")
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
# Настройки вывода 
pd.set_option('display.max_colwidth', None)

# Вспомогательные функции проверок и преобразований
def in_bracets_trim(text):                                            # Удаление скобок и их  содержимого
    # Удаление содержимого в скобках и самих скобок
    while '(' in text and ')' in text:
        text = re.sub(r'\([^()]*\)', '', text)
    return text

def remove_spaces_and_parentheses(text):                          # Убрать пробелы и скобки
    # Удаление содержимого в скобках и самих скобок  
    text = in_bracets_trim(text)
    # Удаление пробелов
    text = text.replace(" ", "")
      
    return text

# Функции преобразований
def city_extraсt(raw_str):                                        # Выделим признак города
    city = raw_str.split(' , ')[0]
    city_filtered = remove_spaces_and_parentheses(city)

    million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    if city_filtered in million_cities: return "город-миллионник"
    elif city_filtered in ["Москва", "Санкт-Петербург"]: return city_filtered
    else: return 'другие'

def ready_to_dep(raw_str):
    lst = raw_str.split(' , ')
    if lst[1].startswith(" м."): lst = lst[2:]
    else: lst = lst[1:]
    str_list = lst[0].split(' ')
    if str_list[0] == 'не': return False
    else: return True

def ready_to_biztrip(raw_str):
    lst = raw_str.split(' , ')
    if lst[1].startswith(" м."): lst = lst[3:]
    else: lst = lst[2:]
    
    try:
        if len(lst) == 0 : raise ValueError("Список имеет нулевую длинну.")
        filtered_str = lst[-1]
        if filtered_str.startswith('не'): return False
        elif filtered_str.startswith('г'): return True
        else: raise ValueError("Ошибка преобразования, строка начинается иначе чем: 'не','г'\n Строка: {filtered_str}    ")
    
    except ValueError as e:
        #print(f"Проблема преобразования\n Значения:{lst}\n Изначальная строка: {raw_str}\n Ошибка: {e}")
        return False

# Блок преобразований
main_df["Город"] = main_df["Город, переезд, командировки"].apply(city_extraсt)
main_df["Готовность к переезду"] = main_df["Город, переезд, командировки"].apply(ready_to_dep)
main_df["Готовность к командировкам"] = main_df["Город, переезд, командировки"].apply(ready_to_biztrip)

# Блок проверки данных
"""display(main_df[["Город, переезд, командировки","Готовность к командировкам"]][ main_df["Готовность к командировкам"]=='г' ])
display(main_df[["Город, переезд, командировки","Готовность к командировкам"]])
print(main_df["Готовность к командировкам"].unique())"""

# Блок вывода статистик
spb_percentage = main_df[main_df["Город"]=='Санкт-Петербург'].shape[0] / main_df.shape[0]

print(f"Проент жителей Санкт-Петербурга: {round(spb_percentage, 2) * 100}")
```

Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
exchange_df = pd.read_csv('ExchangeRates.csv', sep=',')
df1 = main_df

curr_df = exchange_df[["date","currency","proportion","close"]]
curr_df['date'] = pd.to_datetime(curr_df['date'], format='%d/%m/%y')


curr_dict = {
    'руб.': 'RUB',
    'бел.руб.': 'BYN',
    'KZT': 'KZT',  
    'EUR': 'EUR',
    'USD': 'USD',
    'грн.': 'UAH',
    'сум': 'UZS',
    'KGS': 'KGS',
    'AZN': 'AZN'
}
   

df1["currency"] = df1["ЗП"].apply(lambda x: x.split(' ')[1]).map(curr_dict)                           #['руб.', 'бел.руб.', 'KZT', 'EUR', 'USD', 'грн.', 'сум', 'KGS','AZN']  =>  ['RUB', 'BYN', 'KZT', 'EUR', 'USD', 'UAH', 'UZS', 'KGS', 'AZN']
df1["Сумма в валюте ЗП"] = df1["ЗП"].apply(lambda x: x.split(' ')[0]).astype(int)

df1['date'] = df1['Обновление резюме'].apply(lambda x: str(x).split(' ')[0])
df1['date'] = pd.to_datetime(df1['date'], format='%d.%m.%Y')

df1['currency'] = df1['currency'].astype(str)
curr_df['currency'] = curr_df['currency'].astype(str)

merged_df = pd.merge(main_df, curr_df, on=['date', 'currency'], how='left')

merged_df.loc[merged_df['currency'] == 'RUB', 'proportion'] = 1
merged_df.loc[merged_df['currency'] == 'RUB', 'close'] = 1

merged_df["ЗП в руб"] = merged_df.apply(lambda row: row['Сумма в валюте ЗП']*row['close']/row['proportion'], axis=1)

main_df["ЗП в руб"] = merged_df["ЗП в руб"]

#display(main_df[ main_df["currency"]!="RUB" ] )

# Блок вывода статистик 

print(f"Медианная заработная плата: {main_df['ЗП в руб'].median()}")
```
